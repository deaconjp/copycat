###**1. UNDERSTANDING THE DISTRIBUTION OF DATA**
######**HOW ARE THE TERRORIST ATTACKS DISTRIBUTED ACROSS DIFFERENT YEARS?**

```python
attacks_per_year = td['iyear'].value_counts().sort_index()
```

```python
# Plot the trend using a line graph
plt.figure(figsize=(12, 6))
plt.plot(attacks_per_year.index, attacks_per_year.values, marker='o', linestyle='-', color='b')
plt.title('Number of Terrorist Attacks per Year')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

######**WHAT IS THE MONTHLY DISTRIBUTION OF TERRORIST ATTACKS?**

---



```python
attacks_per_month = td['imonth'].value_counts().sort_index()
```

```python

# Plot the distribution using a bar graph
plt.figure(figsize=(12, 6))
plt.bar(attacks_per_month.index, attacks_per_month.values, color='c', edgecolor='black')
plt.title('Number of Terrorist Attacks per Month')
plt.xlabel('Month')
plt.ylabel('Number of Attacks')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Plot the distribution using a bar graph
plt.figure(figsize=(12, 6))
plt.bar(attacks_per_day.index, attacks_per_day.values, color='m', edgecolor='black')
plt.title('Number of Terrorist Attacks per Day of the Month')
plt.xlabel('Day of the Month')
plt.ylabel('Number of Attacks')
plt.xticks(range(1, 32))  # Days of the month can range from 1 to 31
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Show the plot
plt.show()
```

######**WHAT ARE THE MOST COMMON COUNTRIES AND REGIONS AFFECTED BY TERRORISM?**

```python
country_col = 'country'  # Replace with the actual column name for countries
region_col = 'region'    # Replace with the actual column name for regions
```

```python
# Count the number of attacks per country
attacks_per_country = td[country_col].value_counts()

# Count the number of attacks per region
attacks_per_region = td[region_col].value_counts()

```

```python
# Plotting the top 10 regions
plt.subplot(1, 2, 2)
plt.bar(top_10_regions.index, top_10_regions.values, color='r', edgecolor='black')
plt.title('Top 10 Regions Affected by Terrorism')
plt.xlabel('Region')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
# Print the most frequent country and region
most_frequent_country = top_10_countries.idxmax()
most_frequent_region = top_10_regions.idxmax()
print(f"The country with the most frequent attacks is: {most_frequent_country}")
print(f"The region with the most frequent attacks is: {most_frequent_region}")
```

###**3. ANALYZING TARGETS AND IMPACT**



```python
# Aggregate the data by target type:
target_counts = td['targtype1_txt'].value_counts()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x=target_counts.index, y=target_counts.values, palette='viridis')
plt.xticks(rotation=90)
plt.xlabel('Target Type')
plt.ylabel('Frequency')
plt.title('Distribution of Different Target Types in Terrorist Attacks')
plt.show()
```

```python
#Impact of Terrorist Attacks in Terms of Fatalities and Injuries
#Aggregate data by year:

yearly_impact = td.groupby('iyear')[['nkill', 'nwound']].sum().reset_index()
```

```python
plt.figure(figsize=(14, 8))
width = 0.4  # Width of the bars
p1 = plt.bar(yearly_impact['iyear'] - width/2, yearly_impact['nkill'], width, label='Fatalities', color='red')
p2 = plt.bar(yearly_impact['iyear'] + width/2, yearly_impact['nwound'], width, label='Injuries', color='blue')

plt.xlabel('Year')
plt.ylabel('Number of Casualties')
plt.title('Number of Fatalities and Injuries Across Different Years')
plt.legend()
plt.show()
```

```python
#Analyze the number of casualties (fatalities and injuries) for each attack type.
#Aggregate data by attack type:
attack_casualties = td.groupby('attacktype1_txt')[['nkill', 'nwound']].sum().reset_index()
```

```python
attack_casualties_melted = attack_casualties.melt(id_vars='attacktype1_txt', value_vars=['nkill', 'nwound'], var_name='Casualty Type', value_name='Count')

plt.figure(figsize=(14, 8))
sns.barplot(x='attacktype1_txt', y='Count', hue='Casualty Type', data=attack_casualties_melted, palette='viridis')
plt.xticks(rotation=90)
plt.xlabel('Attack Type')
plt.ylabel('Number of Casualties')
plt.title('Number of Casualties Across Different Attack Types')
plt.show()
```

```python
#Extent of Property Damage Caused by Terrorist Attacks
property_damage = td.groupby('attacktype1_txt')['propvalue'].sum().reset_index()
```

```python
plt.figure(figsize=(14, 8))
sns.barplot(x='attacktype1_txt', y='propvalue', data=property_damage, palette='viridis')
plt.xticks(rotation=90)
plt.xlabel('Attack Type')
plt.ylabel('Estimated Property Damage')
plt.title('Estimated Property Damage Across Different Attack Types')
plt.show()
```

###**6. ANALYZING WEAPONS AND ATTACK METHODS**

```python
# Count attacks by weapon type
weapon_counts = td['weaptype1_txt'].value_counts()
```

```python
plt.figure(figsize=(14, 8))
sns.barplot(x=weapon_counts.index, y=weapon_counts.values, palette='viridis')
plt.xticks(rotation=90)
plt.xlabel('Weapon Type')
plt.ylabel('Number of Attacks')
plt.title('Distribution of Different Weapon Types Used in Terrorist Attacks')
plt.show()
```

```python
# Group by weapon type and success, and count the number of attacks
weapon_success = td.groupby(['weaptype1_txt', 'success']).size().reset_index(name='attack_count')
```

```python
# Pivot table to structure data for comparison
success_data = weapon_success.pivot(index='weaptype1_txt', columns='success', values='attack_count').fillna(0)
success_data.columns = ['Failure', 'Success']
```

```python
success_data.plot(kind='bar', stacked=True, figsize=(14, 8), color=['red', 'green'])
plt.xlabel('Weapon Type')
plt.ylabel('Number of Attacks')
plt.title('Success Rates of Attacks by Weapon Type')
plt.legend(title='Attack Outcome')
plt.xticks(rotation=90)
plt.show()
```

```python
# Create a new column to count the number of weapon types used in an attack
td['num_weapons'] = td[['weaptype1', 'weaptype2', 'weaptype3', 'weaptype4']].notna().sum(axis=1)
```

```python
# Count how many attacks involved multiple weapons
weapon_usage_counts = td['num_weapons'].value_counts()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=weapon_usage_counts.index, y=weapon_usage_counts.values, palette='viridis')
plt.xlabel('Number of Weapon Types Used')
plt.ylabel('Number of Attacks')
plt.title('Frequency of Multiple Weapon Usage in Terrorist Attacks')
plt.show()
```
